```python
import tabula
import pandas as pd
```

```python
caminho_E1_P1 = "/content/SP24_1_P1_PIT.pdf"
caminho_E1_P2 = "/content/SP24_1_P2_PITS.pdf"
caminho_E2_P1 = "/content/SP24_2_P1_PITS.pdf"
caminho_E3_P1 = "/content/SP24_3_P1_PITS.pdf"
caminho_E3_P2 = "/content/SP24_3_P2_PIT.pdf"
caminho_E4_P1 = "/content/SP24_4_P1_PIT.pdf"
caminho_E4_P2 = "/content/SP24_4_P2_PIT.pdf"
```

```python
tabela_E1_P1 = tabula.read_pdf(caminho_E1_P1, pages="all", multiple_tables=True, guess=True, stream=True)
tabela_E1_P2 = tabula.read_pdf(caminho_E1_P2, pages="all", multiple_tables=True, guess=True, stream=True)
tabela_E2_P1 = tabula.read_pdf(caminho_E2_P1, pages="all", multiple_tables=True, guess=True, stream=True)
tabela_E3_P1 = tabula.read_pdf(caminho_E3_P1, pages="all", multiple_tables=True, guess=True, stream=True)
tabela_E3_P2 = tabula.read_pdf(caminho_E3_P2, pages="all", multiple_tables=True, guess=True, stream=True)
tabela_E4_P1 = tabula.read_pdf(caminho_E4_P1, pages="all", multiple_tables=True, guess=True, stream=True)
tabela_E4_P2 = tabula.read_pdf(caminho_E4_P2, pages="all", multiple_tables=True, guess=True, stream=True)
```

```python
tabela_E1_P1 = tabela_E1_P1[0]
tabela_E1_P2 = tabela_E1_P2[0]
tabela_E2_P1 = tabela_E2_P1[0]
tabela_E3_P1 = tabela_E3_P1[0]
tabela_E3_P2 = tabela_E3_P2[0]
tabela_E4_P1 = tabela_E4_P1[0]
tabela_E4_P2 = tabela_E4_P2[0]
```

```python
dataframe_E1_P1 = pd.DataFrame(tabela_E1_P1)
dataframe_E1_P2 = pd.DataFrame(tabela_E1_P2)
dataframe_E2_P1 = pd.DataFrame(tabela_E2_P1)
dataframe_E3_P1 = pd.DataFrame(tabela_E3_P1)
dataframe_E3_P2 = pd.DataFrame(tabela_E3_P2)
dataframe_E4_P1 = pd.DataFrame(tabela_E4_P1)
dataframe_E4_P2 = pd.DataFrame(tabela_E4_P2)
```

```python
columns_for_index = ['No.', 'Name', 'Modelo', 'Equipe', 'Lap',  'PT Tm','Time of Day']
```

```python
dataframe_E4_P1.columns = columns_for_index
dataframe_E4_P2.columns = columns_for_index
```

```python
colunas_excluir = ['Modelo', 'Time of Day']
```

```python
dataframe_E1_P1.drop(columns=colunas_excluir, inplace=True)
dataframe_E1_P2.drop(columns=colunas_excluir, inplace=True)
dataframe_E2_P1.drop(columns=colunas_excluir, inplace=True)
dataframe_E3_P1.drop(columns=colunas_excluir, inplace=True)
dataframe_E3_P2.drop(columns=colunas_excluir, inplace=True)
dataframe_E4_P1.drop(columns=colunas_excluir, inplace=True)
dataframe_E4_P2.drop(columns=colunas_excluir, inplace=True)
```

```python
dataframe_E1_P1['Etapa'] = 'Goiânia-GO'
dataframe_E1_P2['Etapa'] = 'Goiânia-GO'
dataframe_E2_P1['Etapa'] = 'Velocitta-SP'
dataframe_E3_P1['Etapa'] = 'São Paulo-SP'
dataframe_E3_P2['Etapa'] = 'São Paulo-SP'
dataframe_E4_P1['Etapa'] = 'Cascavel-PR'
dataframe_E4_P2['Etapa'] = 'Cascavel-PR'
```

```python
df_pitstops = pd.concat([dataframe_E1_P1, dataframe_E1_P2, dataframe_E2_P1, dataframe_E3_P1, dataframe_E3_P2, dataframe_E4_P1, dataframe_E4_P2], ignore_index=True)
```
